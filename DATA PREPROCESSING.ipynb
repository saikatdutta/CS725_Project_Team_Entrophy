```python
def reorder_columns(df, target_column_name, target_position):
    columns = df.columns.tolist()
    columns.insert(target_position, columns.pop(columns.index(target_column_name)))
    return df[columns]
```

```python
import pandas as pd

# Assuming you have your data in a DataFrame called df
# If not, you can read your CSV file using pd.read_csv('your_file.csv')

# Initialize global variables
current_over = None
ball_count = 0

# Function to generate dynamic over numbers
def generate_over(row):
    global current_over, ball_count
    if row['Over'] != current_over:
        current_over = row['Over']
        ball_count = 1
    else:
        ball_count += 1
    return f"{current_over}.{ball_count}"

# Apply the function to generate dynamic over numbers
df['Over'] = df.apply(generate_over, axis=1)

# Display the modified DataFrame
print(df)
```

```python
# Initialize a counter
counter = 1

# Create a list to store the values for the new column
new_column_values = []

# Iterate through the DataFrame
for i in range(len(df)):
    if i > 0 and df.at[i, 'Team'] != df.at[i - 1, 'Team']:
        counter += 1
    new_column_values.append(counter)

# Add the new column to the DataFrame
df.insert(0, 'Innings', new_column_values)

# Display the modified DataFrame
print(df)
```
