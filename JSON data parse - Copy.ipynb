```python
# CODE TO IMPORT LIST OF JSON FILENAMES FROM T20S_JSON DIRECTORY
import os

# Specify the directory path
directory_path = os.getcwd() + '\\t20s_json\\'

# Initialize an empty list to store JSON filenames
json_files = []
json_file_path = []

# Iterate through the files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".json"):
        json_files.append(filename)
        json_file_path.append(directory_path + filename)

# Print the NUMBER OF ITEMS IN list of JSON filenames
print(len(json_files))
print(len(json_file_path))
print(json_file_path[0])
```

```python
# CODE TO CONVERT ALL THE JSON FILES TO CSV FILES

import json
import csv
import os

for i in range(len(json_file_path)):
    # Load the JSON data
    file_path = json_file_path[i]
    if not os.path.exists('t20s_csv'):
        os.mkdir('t20s_csv')
    csv_file_path = os.getcwd() + '\\t20s_csv\\' + os.path.splitext(json_files[i])[0] + ".csv"
    
    with open(file_path, "r") as json_file:
        data = json.load(json_file)
    
    # Extract the innings data
    innings_data = data.get("innings", [])
    
    # Extract and write the data to a CSV file
    with open(csv_file_path, "w", newline="") as csv_file:
        csv_writer = csv.writer(csv_file)
    
        # Write the header row
        csv_writer.writerow(["Team", "Over", "Batter", "Bowler", "Non_Striker", "Batter_Runs", "Extras", "Total_Runs"])
    
        # Write the data rows
        for inning in innings_data:
            for over in inning.get("overs", []):
                for delivery in over.get("deliveries", []):
                    batter_name = delivery["batter"]
                    bowler_name = delivery["bowler"]
                    non_striker_name = delivery["non_striker"]
                    batter_runs = delivery["runs"]["batter"]
                    extras = delivery["runs"]["extras"]
                    total_runs = delivery["runs"]["total"]
                    csv_writer.writerow([inning["team"], over["over"], batter_name, bowler_name, non_striker_name, batter_runs, extras, total_runs])
                    

    print(f"CSV file '{csv_file_path}' has been created.")
```
